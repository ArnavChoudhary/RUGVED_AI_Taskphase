```python
#find the city with the maximum and minimum number of matches held there. 

H = data1.city.value_counts().nlargest(1)
L = data1.city.value_counts().nsmallest(1)
print("City with the highest number of matches is: ", H)
print("City with the lowest number of matches is: ", L)
```

```python
#find total count of matches city wise

T = data1.city.value_counts()
print("The number of matches in different cities is as follows:\n ", T)
```

```python
#find the Team which is maximum and minimum toss winner.

A = data1.toss_winner.value_counts().nlargest(1)
B = data1.toss_winner.value_counts().nsmallest(1)
print("Team with the highest number of toss wins is:",A )
print("Team with the highest number of toss wins is:",B)

```

```python
#check the toss decision that the team has taken.

pd.crosstab(data1['toss_winner'], data1['toss_decision'])
```

```python
#count the total number of normal and tie matches

C = data1.result.value_counts().head(2)
print("The total number of normal and tie matches are as follows:\n",C)
```

```python
#find the team names where the match result is tie
pd.crosstab(data1['toss_winner'], data1['result']=='tie')
```

```python
#find the team name who won the match by highest runs

D = data1['winner'].loc[data1['win_by_runs'].idxmax]
print("The team to win by maximum runs is:",D)
```

```python
#find the team name who won the match by lowest runs

D = data1['winner'].loc[data1['win_by_runs'].idxmin]
print("The team to win by lowest runs is:",D)
```

```python
#find the players who was awarded “Player of the match” more than 3 times.

score = 0
for index, row in data1.iterrows():
    if(row['win_by_runs'] == score):
        print(row['winner'])
```
